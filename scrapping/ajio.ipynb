```python
reviews=soup.find_all('div',{'class':'ZmyHeo'})
for review in reviews:
    print(review.text)
    print('------------------')
```

```python
rating=soup.find_all('div',class_='XQDdHH Ga3i8K')
for review in rating:
    print(review.text)
    print('------------------')
```

```python
names=soup.find_all('p',class_='z9E0IG')
for name in names:
    print(name.text)
    print('------------------')
```

```python
commetTags=soup.findAll('div',{'class':'wys2hv _43gOsC'})
for commetTag in commetTags:
    print(commetTag.get_text())
    answer=commetTag.findNext('div',{'class':'JxAXcP'}).get_text()
    print(answer)
    print('------------------')
```
