```python
import requests
import bs4
```

```python
request1 = requests.get('https://www.shopclues.com/signatize-wireless-in-ear-neckband-bluetooth-earphones-with-10mm-drivers-rich-music-experience-50-hours-playtime-type-153640094.html')
request1
```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
soup
```

### COMMENTS/REVIEWS

```python
reviews = soup.find_all('div', {'class': 'review_desc'})
for review in reviews:
    print(review.get_text() + "\n\n")
```

### RATINGS

```python
ratings = soup.find_all('div', {'class': 'prd_ratings'})
for rating in ratings:
    print(rating.get_text() + "  ")
```

```python
overall_rating = soup.find('div', {'class': 'star_rating_point'}).get_text()
print(overall_rating)
```

### COMMENT TAGS

```python
tags = soup.find_all('div', {'class': 'r_date'})
for tag in tags:
    print(tag.get_text() + "  ")
```

### CUSTOMER NAMES

```python
customers = soup.find_all('div', {'class': 'r_by'})
for customer in customers:
    print(customer.get_text() + "  ")
```
