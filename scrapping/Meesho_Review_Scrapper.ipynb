```python
import requests
import bs4
```

```python
request1 = requests.get('https://www.meesho.com/tusi-tank-top-combo-black-white-01/p/3x8t00')
request1
```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
soup
```

### COMMENTS/REVIEWS

```python
reviews = soup.find_all('span', {'class': 'sc-eDvSVe gUjMRV Comment__CommentText-sc-1ju5q0e-3 cfdxfJ Comment__CommentText-sc-1ju5q0e-3 cfdxfJ'})
for review in reviews:
    print(review.get_text() + "\n\n")
```

### RATINGS

```python
ratings = soup.find_all('span', {'class': 'sc-eDvSVe laVOtN'})
for rating in ratings:
    print(rating.get_text() + "  ")
```

```python
overall_rating = soup.find('h1', {'class': 'sc-eDvSVe cdZTwf'}).get_text()
print(overall_rating)
```

### COMMENT TAGS

```python
tags = soup.find_all('span', {'class': 'sc-eDvSVe XndEO'})
for tag in tags:
    print(tag.get_text() + "  ")
```

### CUSTOMER NAMES

```python
customers = soup.find_all('span', {'class': 'sc-eDvSVe dugLmN'})
for customer in customers:
    print(customer.get_text() + "  ")
```
